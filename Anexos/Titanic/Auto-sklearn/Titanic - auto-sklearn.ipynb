This is the notebook of using [auto-sklearn](https://github.com/automl/auto-sklearn), for the challenge of data insensitive analytics.

By using auto-sklearn, you need to care the following things only!
- is the problem regression or classification?
  - using `AutoSklearnClassifier` or `AutoSklearnRegressor`?
- is the each column `Categorical` or `Numerical`?

```python
import pandas as pd
pd.set_option("display.max_colwidth", 80)
pd.set_option("display.max_columns", 80)
pd.set_option("display.max_rows", 6)
```

```python
df = pd.read_csv('/dbfs/titanic/train.csv')
df
```

```python
# specify the column type, categorical or numerical, for each column
C = 'Categorical'
N = 'Numerical'
label_name = 'Survived'
feature_dict = {
    'Pclass': C,
    'Sex': C,
    'Age': N,
    'SibSp': C,
    'Parch': C,
    'Ticket': C,
    'Fare': N,
    'Cabin': C,
    'Embarked': C
}
```

```python
from sklearn import preprocessing

def fillna(df):
    na_list = [i for i in feature_dict.keys() if df[i].isnull().values.any()]
    for i in na_list:
        if feature_dict[i] == N:
            df[i] = df[i].fillna(value=df[i].mean())
        else:
            df[i] = df[i].fillna(method='backfill')
            df[i] = df[i].fillna(method='pad')
            # df[i] = df[i].fillna(value=df[i].value_counts().index[0])

categorical_names = list(map(lambda x: x[0], filter(lambda x: x[1] == C, feature_dict.items())))
def categoricalize(df):
    le = preprocessing.LabelEncoder()
    for column in categorical_names:
        # print((column, df[column].shape))
        df[column] = le.fit_transform(df[column])
```

```python
fillna(df)
df
```

```python
categoricalize(df)
df
```

```python
features = df[list(feature_dict.keys())]
feat_type = list(feature_dict.values())
labels = df[label_name]
```

```python
import autosklearn.classification
import sklearn.model_selection
import sklearn.metrics

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(features.as_matrix(), labels.as_matrix(), random_state=1)
```

```python
y_test
```

```python
automl = autosklearn.classification.AutoSklearnClassifier(
  time_left_for_this_task=120
  #   time_left_for_this_task=120, per_run_time_limit=30
)
```

```python
automl.fit(X_train.copy(), y_train.copy(), feat_type=feat_type, dataset_name='titanic')
print(automl.show_models())
predictions = automl.predict(X_test)
print("Accuracy score", sklearn.metrics.accuracy_score(y_test, predictions))
```

```python
test_df = pd.read_csv('/dbfs/titanic/test.csv')[list(feature_dict.keys())]
test_df
```

```python
fillna(test_df)
categoricalize(test_df)
```

```python
predictions = automl.predict(test_df)
```

```python
preds = pd.DataFrame({'PassengerId' : range(892,1310), 'Survived' : pd.Series(predictions)})
preds.to_csv('/dbfs/titanic/results-autosklearn.csv', index=False)
```
