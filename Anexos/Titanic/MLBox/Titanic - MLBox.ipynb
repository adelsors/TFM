```
rd = Reader(sep = ",")
df = rd.train_test_split(paths, target_name)   #reading and preprocessing (dates, ...)
```

```
dft = Drift_thresholder()
df = dft.fit_transform(df)   #removing non-stable features (like ID,...)
```

```
opt = Optimiser(scoring = "accuracy", n_folds = 5)
```
