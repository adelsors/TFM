```python
import pandas as pd
import openml
from sklearn.ensemble import RandomForestClassifier
from supervised.automl import AutoML

import os
import numpy as np
import pandas as pd
import sklearn.model_selection
from sklearn.metrics import log_loss, f1_score
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./train.csv')
df.head(5)
```

```python
x_cols = [c for c in df.columns if c != 'target']
X = df[x_cols]
y = df['target']
```

```python
X.head(5)
```

```python
%%time 

a = AutoML(total_time_limit=12*60*60, 
           tuning_mode='Perfect', 
           ml_task='binary_classification', 
           explain_level = 2) 
```

```python
%%time

a.fit(X, y)
```

```python
df_test = pd.read_csv('./test.csv')
df_test.head(5)
```

```python
%%time 

df_pred = a.predict(df_test)
df_pred.head(5)
```

```python
#save predict results to submission form
submit = pd.read_csv("./sample_submission.csv",sep=',')

submit['target'] =  df_pred["prediction_1"].values

submit.to_csv("./results-mljar-supervised-20200602-orig.csv", index=False)
```
